```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# 1. CSV dosyalarını oku
df_lemma = pd.read_csv("vectors/tfidf_lemmatized_5000.csv")
df_stem = pd.read_csv("vectors/tfidf_stemmed_5000.csv")

# 2. Tüm kelimeleri bir listeye topla
words_lemma = []
for col in df_lemma.columns:
    words_lemma.extend([col] * int(df_lemma[col].astype(bool).sum()))

words_stem = []
for col in df_stem.columns:
    words_stem.extend([col] * int(df_stem[col].astype(bool).sum()))

# 3. Kelime frekanslarını hesapla
freq_lemma = Counter(words_lemma)
freq_stem = Counter(words_stem)

# 4. Sıralı frekans listeleri
lemma_freq_sorted = sorted(freq_lemma.values(), reverse=True)
stem_freq_sorted = sorted(freq_stem.values(), reverse=True)

# 5. Zipf grafiğini çiz
def plot_zipf(frequencies, title):
    ranks = np.arange(1, len(frequencies) + 1)
    plt.figure(figsize=(8, 5))
    plt.loglog(ranks, frequencies)
    plt.xlabel("Kelime Sırası (Rank) [log]")
    plt.ylabel("Frekans [log]")
    plt.title(title)
    plt.grid(True)
    plt.show()

plot_zipf(lemma_freq_sorted, "Zipf Grafiği - Lemmatized Veriler (5000 Cümle)")
plot_zipf(stem_freq_sorted, "Zipf Grafiği - Stemmed Veriler (5000 Cümle)")
```
